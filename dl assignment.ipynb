```python
import pandas as pd
import numpy as np
train=pd.read_csv("train.csv")
train.head()
```

```python
train=train.drop(columns=["id"])
x_train=train.iloc[:,:-1]
y_train=train.iloc[:,-1:]
x_train
```

```python
test=pd.read_csv("test.csv")
test.head()
```

```python
test=test.drop(columns=["id"])
x_test=test.iloc[:,:]
x_test
```

```python
model=keras.Sequential() 
model.add(keras.layers.Dense(100,input_dim=16,activation='relu')) 
model.add(keras.layers.Dense(300,input_dim=100,activation='relu')) 
model.add(keras.layers.Dense(400,input_dim=300,activation='relu')) 
model.add(keras.layers.Dense(500,input_dim=400,activation='relu'))
model.add(keras.layers.Dense(600,input_dim=500,activation='relu')) 
model.add(keras.layers.Dense(1,input_dim=700)) 
```

```python
model.compile(optimizer='adam',loss="mean_squared_error",metrics="mse") 
```

```python
model.fit(x_train,y_train,epochs=30) 
```

```python
y_pred=model.predict(x_train)
```

```python
from sklearn.metrics import r2_score 
r2_score(y_pred,y_train) 
```
